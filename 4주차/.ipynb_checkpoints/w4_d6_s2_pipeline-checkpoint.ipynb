# Pipeline 사용 목적
- 실제 연속되는 전처리와 모델 간을 한꺼번에 학습할 수 있음(전처리와 알고리즘을 합쳐서) -> 전처리 + 모델의 영향력을 한 번에 알 수 있음
- 단점: 일방적으로 전체 데이터를 다 사용해야 하기 때문에 컬럼별 특성을 고려 못함
- 그래서 컬럼 트랜스토멀한 애를 사이킷 .2에 지원하기 시작

# 만드는 방식
- 객체 방식(인스턴스화 해서 만드는 방식)
- 팩토리 메소드

```python
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.iloc[:,:-1], iris.species)
```

```python
ss = StandardScaler()
```

```python
ss.fit_transform(iris[['sepal_length']]) # 팬시 인덱싱
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
pipe = Pipeline([('ss', StandardScaler()),('knn', KNeighborsClassifier())])
# 인스턴스화해서 만드는 방식
# s+t # step: ...마지막에 estimator를 넣어야 함
# numpy에서 온 기법 # structured array 할 때, 리스트로, 그 안에 튜플로...
# 맨 마지막 에스티메이터를 가지고만 있으면 튜플을 몇 개를 넣든 상관 없음
# 왜 에스티메이터?
# 전처리와 에스티메이트를 함께 고려해서 성능 측정하기 위해서 하는 거니까
# 이름과, 인스턴스한 애
```

```python
vars(pipe) # 가지는 인스턴스 값들을 표현
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.score(X_test, y_test)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.drop(columns='name', inplace=True)
```

```python
mpg
```

```python
ss = StandardScaler()
```

```python
ss.fit_transform(mpg)
```

```python
# origin이 문자열이라서 바꿀 수 없음 -> 파이프라인을 쓸 수 없다
# 이럴 때 컬럼트랜스포머를 쓰는데, 핵심은 여러 개의 파이프라인을 붙인다는 것
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
pipe1 = Pipeline([('imputer', SimpleImputer())])
pipe2 = Pipeline([('la', LabelEncoder()), ('ss', StandardScaler())])
pipe3 = Pipeline([('ss', StandardScaler())])
# origin에 대해서 라벨 인코딩 # 나머지 데이터에 대해서 스탠다드스케일러
```

```python
ct = ColumnTransformer([
    ('im', pipe1, 'horsepower'), # 이름, 트랜스포머, 적용시킬 컬럼
    ('l', pipe2, 'origin'), # pandas와 연동된다는 뜻
    ('s', pipe3, ['cylinders', 'displacement'])
])
```

```python
pipe = Pipeline([('ct', ct), ('knn', KNeighborsClassifier())])
# 컬럼트랜스포머 안에 파이프라인을 넣음
```

```python
pipe.fit(X_train, y_train)
```

```python
ct = ColumnTransformer([
    ('im', pipe1, mpg.horsepower), 
    ('l', pipe2, mpg.origin),
    ('s', pipe3, [mpg.cylinders, mpg.displacement])
])
```

```python
pipe = Pipeline([('ct', ct), ('knn', KNeighborsClassifier())])
```

```python
pipe.fit(X_train, y_train)
```
