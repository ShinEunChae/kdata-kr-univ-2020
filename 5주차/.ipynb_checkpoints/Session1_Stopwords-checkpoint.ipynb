```python
from konlpy.corpus import kobill
corpus = ''.join([kobill.open(_).read() for _ in kobill.fileids()])
```

```python
from nltk.tokenize import word_tokenize
```

```python
voca = word_tokenize(corpus)
```

```python
L = dict()
R = dict()
```

```python
for term in voca:
    if len(term) > 1:
        for i in range(1, len(term)-1):
            lkey = term[:i+1]
            rkey = term[i+1:]
            if lkey in L.keys():
                L[lkey] += 1
            else:
                L[lkey] = 0
            if rkey in R.keys():
                R[rkey] += 1
            else:
                R[rkey] = 0
```

```python
sorted(L.items(), key=lambda _:_[1], reverse=True)[:10]
```

```python
sorted(R.items(), key=lambda _:_[1], reverse=True)[:10]
```

```python
s = "가수 에릭남이 '철파엠'에 출연해 입담을 뽐냈다."
s = re.sub('[{}]'.format(re.escape(punctuation)), '', s)
```

```python
s
```

```python
for term in word_tokenize(s):
    if len(term) > 1:
        print(term)
        prob = dict()
        for i in range(1, len(term)-1):
            rkey = term[i+1:]
            if rkey in R.keys():
                prob[rkey] = R[rkey]
        total = sum(prob.values())
        if len(prob.keys()) == 0:
            print('명사추정')
        else:
            maxkey = max(prob, key=prob.get)
            print({k:v/total for k, v in prob.items()})
            print(re.sub(maxkey, '', term))
        print()
```

```python
stopwords = ['씨발', '시발']
s = '아 시발X 짜증나'
_s = list()
for _ in word_tokenize(s):
    if _ in stopwords:
        _s.append('*'*len(_))
    else:
        _s.append(_)
print(' '.join(_s))
```

```python
stopwords = ['씨발', '시발', '포장']
s = '아 포장 선물포장 씨!$%^1발 ㅆ1발 짜증나'
s = re.sub('[{}]'.format(re.escape(punctuation)), '', s)
s = re.sub('\B[0-9]\B', '', s)
s = re.sub('[ㄱ-ㅎㅏ-ㅣ]', '', s)
_s = list()
for _ in word_tokenize(s):
    if re.search(r'({})'.format('|'.join(stopwords)), _):
        _s.append('*'*len(_))
    else:
        _s.append(_)
print(' '.join(_s))
```

```python
# BPE
def convert_data(data): # {'문자열':빈도}
    newdata = dict()
    for k,v in data.items():
        newdata[' '.join(['<w>'] + list(k) + ['</w>'])] = v
    return newdata

def find_pair(data): # 문자열 -> 쌍으로 반환
    pair = defaultdict(lambda:0)
    for k, v in data.items(): # l o w </w> : 5
        tokens = k.split() # [l, o, w, </w>]
        for i in range(len(tokens)-1): # (l, o):5, (o, w):5 # (l, o) 또 나올 수 있음
            pair[tuple(tokens[i:i+2])] += v # -> 빈도를 다 더해줌
    return pair

def merge_pair(data, key): # 맥스 패턴 -> 합치기
    newdata = dict()
    for k, v in data.items():
        newkey = re.sub(' '.join(key), ''.join(key), k)
        newdata[newkey] = v
        # 'e s'  이 패턴을 찾겠단 얘기
    return newdata
```

```python
data = convert_data({'시발':5, '씨발':2, '씨-발':6, '씨!@#$발':3})
```

```python
data
```

```python
for _ in range(3):
    pair = find_pair(data)
    key = max(pair, key=pair.get)
    data = merge_pair(data, key)
    
temp = list(set([token for _ in data.keys() for token in _.split()
          if len(token) > 1 and token not in ['<w>', '</w>']]))
```

```python
stop = defaultdict(lambda:0)
for _ in temp:
    for k, v in data.items():
        if re.search(_, k):
            stop[_] += v
stop
```

```python
L = [_ for _ in stop.keys() if re.search('^<w>', _)]
R = [_ for _ in stop.keys() if re.search('</w>$', _)]
```

```python
patterns = list()
for l in L:
    for r in R:
        patterns.append(re.compile('{0}.+?{1}'.format(l, r)))
```

```python
patterns
```

```python
s = '아 포장 선물포장 씨!$%^1발 ㅆ1발 짜증나'
```

```python
_s = list()
for token in s.split():
    skip = False
    for p in patterns:
        if p.search('<w>'+token+'</w>'):
            skip = True
    if skip:
        _s.append('*'*len(token))
    else:
        _s.append(token)
print(' '.join(_s))
```
