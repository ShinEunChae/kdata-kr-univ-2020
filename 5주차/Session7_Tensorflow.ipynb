```python
in_ = tf.keras.layers.Input(shape=(28*28,)) # input data가
layer1 = tf.keras.layers.Dense(64, activation='relu')(in_) # 레이어1에 들어가고, 이 레이어가
layer2 = tf.keras.layers.Dense(64, activation='relu')(in_)
layer3 = tf.keras.layers.Concatenate()([layer1, layer2])
out = tf.keras.layers.Dense(10, activation='relu')(layer3) # out에 들어간다
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train = X_train.reshape(-1, 28*28)
```

```python
model = tf.keras.models.Model(inputs=in_, outputs=out)
```

```python
model.summary()
```

```python
# 자동 차별화 문서 참고
```

```python
from tensorflow.keras.callbacks import Callback, TensorBoard
```

```python
Callback. # 나오는 애들 오버라이딩 가능
Callback.on_epoch_begin
```

```python
class MyCallback(Callback):
    def on_epoch_begin(self, epoch, logs=None):
        print(epoch, 'start!!!')
```

```python
def build_fn():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(16, input_shape=(28*28,), activation='relu'),
        tf.keras.layers.Dense(16, activation='relu'),
        tf.keras.layers.Dense(10, activation='softmax')
        ]
    )
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
model = build_fn()
```

```python
model.fit(X_train, y_train, epochs=5, callbacks=[MyCallback()])
```

```python
model.fit(X_train, y_train, epochs=5, callbacks=[TensorBoard()])
```
